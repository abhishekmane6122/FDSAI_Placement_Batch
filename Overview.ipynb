### **Euclidean Distance**

**Formula**:

$$d(p, q) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}$$

**When to Use**: Euclidean distance is often used when the data is continuous and when you assume that the distance between points is linear and uniform across all dimensions. It is particularly common in clustering algorithms like **K-Means** and in **K-Nearest Neighbors (KNN)** for classification tasks.

**Why & How**:
- **Why**: Euclidean distance works well in low-dimensional spaces where the relationship between data points is approximately linear and the scales of different features are comparable.
- **How**: You calculate Euclidean distance by summing the squared differences for each dimension and then taking the square root. It is simple and intuitive, making it easy to interpret.

**Practical Example**: In image recognition, Euclidean distance is often used to compare pixel intensities between images to determine their similarity. If the images are aligned and the difference between them is mainly in intensity, Euclidean distance will give a good measure of their similarity.

### **Manhattan Distance**
**Definition**: Manhattan distance, also known as L1 distance or city block distance, measures the distance between two points by summing the absolute differences of their Cartesian coordinates.

**Formula**:

$$d(p, q) = \sum_{i=1}^{n} |q_i - p_i|$$

**When to Use**: Manhattan distance is useful when the differences between dimensions are not proportional, or when you are working in high-dimensional spaces where Euclidean distance may become less meaningful. It is commonly used in **LASSO regression** and when dealing with grid-like structures (e.g., in digital image processing or urban planning).

**Why & How**:
- **Why**: It is more robust to outliers and works well in scenarios where only horizontal and vertical movements are allowed, like in a grid layout.
- **How**: You calculate it by summing the absolute differences of each coordinate pair. This is computationally cheaper than Euclidean distance, especially in high-dimensional spaces.

**Practical Example**: In recommendation systems, Manhattan distance can be used to compare the similarity between user preferences or ratings, particularly when ratings are sparse and the differences are likely to be non-linear.

### **Cosine Distance**
**Definition**: Cosine distance measures the cosine of the angle between two vectors in a multi-dimensional space, and it is used to determine the orientation difference between them rather than their magnitude.

**Formula**:

$$d(p, q) = 1 - \cos(\theta) = 1 - \frac{p \cdot q}{\|p\| \|q\|}$$

**When to Use**: Cosine distance is most effective when the magnitude of the vectors is irrelevant, and the focus is on the orientation (i.e., the direction) of the vectors. It is commonly used in **text mining** and **information retrieval** where documents are represented as vectors in a high-dimensional space.

### **BLEU (Bilingual Evaluation Understudy) Score**
**Definition**: BLEU is a metric for evaluating the quality of text that has been machine-translated from one language to another. It is based on comparing the n-grams of the generated translation with those of a reference translation.

**Formula**:

$$BLEU = BP \cdot \exp\left( \sum_{n=1}^{N} w_n \log p_n \right)$$

**When to Use**: BLEU is primarily used in machine translation but can also be used to evaluate other types of text generation tasks in NLP, such as summarization or dialogue systems.

**Why & How**:
- **Why**: BLEU is one of the first and most widely used metrics for automatic evaluation of machine-generated translations because it correlates well with human judgment.

**Practical Example**: BLEU is used to evaluate the performance of neural machine translation models like Google's Transformer. By comparing the output of the model with reference translations, researchers can quantitatively assess the quality of translations.

### **Cross-Entropy**
**Definition**: Cross-entropy is a loss function often used in classification tasks. It measures the difference between two probability distributions, typically the true distribution (labels) and the predicted distribution from the model.

**Formula**:

$$H(p, q) = - \sum_{i=1}^{n} p(x_i) \log q(x_i)$$

**When to Use**: Cross-entropy is widely used as a loss function in classification tasks, including binary, multi-class, and multi-label classification. It is the go-to loss function for training deep neural networks.

**Why & How**:
- **Why**: Cross-entropy provides a measure of how well the predicted probabilities align with the true labels. It is particularly effective because it penalizes incorrect predictions more heavily, encouraging the model to produce more accurate predictions.
- **How**: During model training, the cross-entropy loss is minimized using optimization algorithms like stochastic gradient descent. The loss provides feedback to update the model's parameters.

**Practical Example**: In a neural network trained for image classification, the cross-entropy loss is used to evaluate how well the model's predicted probabilities match the true labels. By minimizing this loss, the model improves its predictive accuracy.
